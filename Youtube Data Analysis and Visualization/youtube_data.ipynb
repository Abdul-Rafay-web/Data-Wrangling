```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Global Youtube Statistics.csv', encoding='latin-1')
```

```python
df.head()
```

```python
pd.set_option('display.float_format','{:.2f}'.format)
```

```python
category_subscribers_mean=df.groupby('category')['subscribers'].mean()
category_subscribers_mean
```

```python
category_subscribers_mean.sort_values(ascending=False)
```

```python
category_subscribers_mean.sort_values(ascending=False).head().plot(kind='bar')
plt.title("Subscribers by Category")
plt.xlabel('Category')
plt.ylabel('Subscribers')
plt.show()
```

```python
channel_type_counts=df['channel_type'].value_counts()
channel_type_counts
```

```python
channel_type_counts.head().plot(kind='pie')
plt.title('Channel Type Distribution')
plt.show()
```

```python
plt.scatter(df['subscribers'],df['video views'])
plt.title('Subscribers vs Video Views')
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.show()
```
